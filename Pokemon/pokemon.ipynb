```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('pokemon.csv')
f, ax= plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt='.1f', ax=ax)
plt.show()
```

```python
df.info()
```

```python
df.Speed.plot(color="g", label="Speed", linewidth=1, alpha=0.5, grid=True, linestyle=":")
df.Defense.plot(color="r", label="Defense", linewidth=1, alpha=0.5, grid=True, linestyle="-.")
plt.legend(loc="best")
plt.xlabel("Pokemons")
plt.ylabel("Gucler")
plt.title("Line Plot")
plt.show()
```

```python
df.plot(kind="scatter", color="r", x="Attack", y="Defense", alpha=0.5)
plt.xlabel("Attack")
plt.ylabel("Attack")
plt.title("Atc-Def Karşılaştırma")
plt.show()
```
